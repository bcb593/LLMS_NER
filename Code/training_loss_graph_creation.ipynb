### Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import pickle
```

### Load Training Logs

```python
with open("training_log_BERT", 'rb') as f1:
    BERT_training_log = pickle.load(f1)
    
with open("training_log_RoBERTa", 'rb') as f2:
    RoBERTa_training_log = pickle.load(f2)

with open("training_log_ALBERT", 'rb') as f3:
    ALBERT_training_log = pickle.load(f3)
    
with open("training_log_GPT2", 'rb') as f4:
    GPT2_training_log = pickle.load(f4)
```

### Read Loss Data

```python
# Step and Loss Lists:
steps = []
loss_BERT = []
loss_RoBERTa = []
loss_ALBERT = []
loss_GPT2 = []

for i in range(len(BERT_training_log) - 1):
    step = BERT_training_log[i]
    steps.append(step["step"])
    loss_BERT.append(step['loss'])
    
    loss_RoBERTa.append(RoBERTa_training_log[i]['loss'])
    loss_ALBERT.append(ALBERT_training_log[i]['loss'])
    loss_GPT2.append(GPT2_training_log[i]['loss'])
```

### Graph Generation

```python
plt.figure(figsize=(10,8))
plt.grid()

plt.plot(steps, loss_BERT)
plt.plot(steps, loss_RoBERTa)
plt.plot(steps, loss_ALBERT)
plt.plot(steps, loss_GPT2)

plt.legend(["BERT", "RoBERTa", "ALBERT", "GPT2"])
plt.title("Model Training Loss vs. Step")
plt.xlabel("Training Step")
plt.ylabel("Loss")
```
